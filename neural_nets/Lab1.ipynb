```python
# Typical import for numpy
# We will use a utility function or two for now...
import numpy as np
```

```python
#Typical import for Matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from math import e

x = np.linspace(-5,5,150)
y = 1/(1+e**(-x))
plt.plot(x,y)
plt.title(r'Plotting of the equation y = $\frac{1}{1+e^{-x}}$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Problem 3

### 3a. Graphing using SymPy: $\;\; \frac{dy}{dx} = \frac{1}{1-e^{-x}}$

```python
# Import sympy tools
from sympy import *
init_printing(use_latex=True)
```

```python
x = symbols('x')
y = 1/(1+e**(-x))
```

```python
dydx = y.diff(x)
```

```python
plot(dydx)
```

### 3b. Graphing using matplotlib: $\;\; \frac{dy}{dx} = \frac{1}{1-e^{-x}}$

```python
#from math import e

dx = np.linspace(-5,5,150)
dy = ((e**(-x))/((1+(e**-x))**2))
plt.plot(dx,dy)
plt.title(r'Graph the derivative of $f(x) = \frac{1}{1+e^{-x}}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
